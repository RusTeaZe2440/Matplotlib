```python
import seaborn

tips = seaborn.load_dataset('tips')
df = tips['day'].replace({'Sat':0,'Sun':1,"Thur":2,"Fri":3})
tips['day'] = df
tips.head()
tips['size'] = tips['size'] * 50
tips.head()

tips['Amount'] = tips['total_bill'] + tips['tip']
tips.head()
```

```python
tips.plot(kind='scatter',x='total_bill',y='tip',marker='o',c=tips['day'],colormap='viridis',s='size',figsize=(8,5))
```

```python
import pandas as pd
batsman = pd.read_csv('batsman_season_record.csv')
batsman
```

```python
batsman.plot(kind='bar')
```

```python
batsman.plot(kind='bar',stacked=True,x='batsman',colormap='viridis_r')
```

```python
batsman[['2015','2016']].plot(kind='hist',colormap='winter')
```

```python
batsman.plot(kind='pie',y='2015',subplots=True)
```

```python
explode = [0,0,0,.1,0]
batsman[['2015','2016','2017']].plot(kind='pie',
                                     subplots=True,
                                     figsize=(20,10),
                                     labels=batsman['batsman'],
                                     explode=explode,
                                     shadow=True
                                     )
```

```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tip = tips.pivot_table(index=['day','time'],columns=['sex','smoker'],values='total_bill',aggfunc='mean',dropna=True)
tip.plot(kind='pie',subplots=True,figsize=(20,15),autopct='%1.1f%%')
```
