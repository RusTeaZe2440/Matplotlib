```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
x = np.linspace(-5,5,10)
y = np.linspace(5,10,10)

X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)

plt.contour(X,Y,Z , levels=20, )
```
