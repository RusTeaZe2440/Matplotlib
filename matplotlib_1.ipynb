* color encoded scatter plot

```python
def create_color_encoded_scatterplot(x:pd.Series,y:pd.Series) -> plt.scatter:
    plt.style.use('ggplot')
    plt.scatter(x,y,marker='o',c=df['day'],edgecolors='black')
    plt.title('Color-Encoded-Scatterplot')
    plt.xlabel('Total-Bill')
    plt.ylabel('Tip')
    plt.colorbar()
    plt.show()

create_color_encoded_scatterplot(df['total_bill'], df['tip'])
    
```

## 3D SCATTER PLOT

```python
z = df['total_bill'] * df['tip']
```

```python

def create_3d_scatter(x:pd.Series,y:pd.Series,z) -> None:
    fig = plt.figure(figsize=(10,10))
    ax = plt.subplot(2,2,1,projection='3d')
    ax.scatter3D(x,y,z,c=df['day'],alpha=1)

create_3d_scatter(df['total_bill'],df['tip'],z)
```

### BAR PLOT

* Bivariate Analysis
* Numerical vs Categorical data
* use case -> Aggregate analysis of groups

```python
children : list[int] = [10,20,30,50,30,10]
colors : list[str] = ['Red', 'Pink','Orange','Yellow','White','Black']

def create_vertical_bar_plot(x:list[int],y:list[str]) -> plt.bar:
    plt.figure(figsize=(8,5))
    plt.bar(x,y,color='purple')
    plt.title('Bar Plot')
    plt.xlabel('Colors')
    plt.ylabel('children')
    plt.show()

create_vertical_bar_plot(children,colors)
```

```python
children : list[int] = [10,20,30,50,30,10]
colors : list[str] = ['Red', 'Pink','Orange','Yellow','White','Black']

def create_horizonta_barplot(children:list[int],colors=list[str]) -> None:
    plt.figure(figsize=(8,5))
    plt.barh(colors,children,color='purple')
    plt.title('Bar Plot')
    plt.xlabel('Colors')
    plt.ylabel('children')
    plt.show()

create_horizonta_barplot(children,colors)
```

```python
# plotting multiple bar plot using matplotlib

def multiple_barplot():
    plt.figure(figsize=(8,5))
    plt.bar(np.arange(ipl.shape[0]) - 0.3 , ipl['2015'],width=0.2,label='2015')
    plt.bar(np.arange(ipl.shape[0]), ipl['2016'],width=0.2,label='2016')
    plt.bar(np.arange(ipl.shape[0]) + 0.3 , ipl['2017'],width=0.2,label='2017')
    plt.xticks(np.arange(ipl.shape[0]),ipl['batsman'])
    plt.ylabel('runs_socred')
    plt.legend()
    plt.show()

multiple_barplot()
```

```python
# stacked bar chart 
def stacked_barplot():
    plt.figure(figsize=(8,5))
    plt.bar(ipl['batsman'], ipl['2017'],label='2017',width=0.5,color='green')
    plt.bar(ipl['batsman'],ipl['2016'],bottom=ipl['2017'],label='2016',width=0.5,color='teal')
    plt.bar(ipl['batsman'],ipl['2015'],bottom=ipl['2017'] + ipl['2016'],label='2015',width=0.5,color='black')
    plt.title("Runs Scored by Batsman in Each Ipl Editon")
    plt.xlabel('Batsman')
    plt.ylabel('Runs Scored')
    plt.legend()
    plt.show()

stacked_barplot()
```

### HISTOGRAM

* Univariate analysis 
* Numerical columns
* Use case - Frequency count

```python
data = np.random.randn(1000)

def create_histogram(data:np.random.randn) -> None:
    plt.hist(data,bins=20,color='#096F63',edgecolor='black',label='Distribution')
    plt.xlabel('Data')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.legend()
    plt.show()
    
create_histogram(data)
```

### PIE CHART

* Univariate / Bivariate Analysis
* Categorical vs Numerical
* Use-Case -> To find contribution on standard scale

```python
# pie chart on simple data
d = [24,55,76,34,67]
subjects= ['History','Science','Maths','English','Geography']
colors = ['Teal','orange','blue','red','purple']
explode = [0.2,0,0.2,0,0] # explode is used to remove slice of pie chart
plt.style.use('ggplot')
plt.pie(d,labels=subjects,explode=explode,shadow=True,colors=colors,autopct='%1.1f%%')
plt.title('Pie chart on simple data')
plt.savefig('sample_piechart.png') # to save file we should remove plt.show command
```
