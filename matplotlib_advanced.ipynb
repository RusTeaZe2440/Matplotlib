```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.randn(1000)

class visualizations:
    def __init__(self,data) -> None:
        self.data = data

    def create_plot(self):
        plt.hist(self.data,bins=30,color='purple',edgecolor='black',label='Distributions')
        plt.title('Simple Histogram')
        plt.xlabel('Frequency')
        plt.ylabel('Data')
        plt.legend()
        plt.show()


v = visualizations(data)
v.create_plot()
```

```python

def create_scatter(x:pd.Series,y:pd.Series) -> plt.scatter:
    plt.scatter(x,y,c=iris['Species'],alpha=1,cmap='summer')
    plt.title('Scatter Plot')
    plt.colorbar()
    plt.xlabel('Sepal Width')
    plt.ylabel('Petal Width')
    plt.show()

create_scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
```

```python
batter = pd.read_csv('batter.csv')
batter.head()
```

```python
sample_df = batter.head(100).sample(25,random_state=5)
sample_df
```

```python


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])


```

```python
# writing axis horizontal line
x = sample_df['avg']
y = sample_df['strike_rate']
z = sample_df['batter']

def create_scatter_labelings(x:pd.Series,y:pd.Series,z:pd.Series) -> plt.scatter:
    for i in range(sample_df.shape[0]):
        plt.text(x.values[i],y.values[i],z.values[i])

    plt.scatter(sample_df['avg'],sample_df['strike_rate'],color='red',edgecolors='black')
    plt.ylabel('Strike-rate')
    plt.xlabel('Average')
    plt.title('Strike-rate vs Avg')
    plt.axhline(125,color='green') ## introduces horizontal line
    plt.axvline(28,color='green') ## introduces vertical line
    plt.show()

create_scatter_labelings(x,y,z)

```

### SUBPLOTS

```python
import matplotlib.pyplot as plt

# Create a 2x1 grid of subplots
fig, ax = plt.subplots(nrows=2, ncols=1,figsize=(8,5))

# Plot something on each subplot
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red',marker='o',edgecolor='black')  # Line plot on the first subplot
ax[0].set_title('Avg Vs Strike-rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike-rate')

ax[1].scatter(batter['runs'],batter['avg'],color='green',edgecolor='black',label='Scatter')  # Scatter plot on the second subplot
ax[1].set_title('Avg Vs Runs')
ax[1].set_xlabel('Runs')
ax[1].set_ylabel('Average')


# Show the plot
plt.show()
```

```python
def create_subplots():
    fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(18,10))

    # Plot something on each subplot

    ax[0, 0].plot(df['batter'],df['runs'],marker='o',linewidth=2.5,color='#E14F0F',linestyle='--',label='Runs')
    ax[0, 0].set_title('Line Plot')
    ax[0, 0].set_xlabel('Batsman')
    ax[0, 0].set_ylabel('Runs')
    ax[0, 0].legend()


    ax[0, 1].scatter(batter['avg'],batter['strike_rate'],color='red',marker='o',edgecolor='black',label='Scatter')
    ax[0, 1].set_title('Scatter Plot')
    ax[0, 1].set_xlabel('Average')
    ax[0, 1].set_ylabel('Strike-Rate')
    ax[0, 1].legend()

    ax[1, 0].bar(df['batter'],df['runs'],color='#13DA90',edgecolor='black',label='bar')
    ax[1, 0].set_title('Bar Graph')
    ax[1, 0].set_xlabel('Batsman')
    ax[1, 0].set_ylabel('Runs')
    ax[1, 0].legend()

    ax[1, 1].hist(sample_df['runs'],bins=25,color='#B30F3F',edgecolor='black',label='Distribution')
    ax[1, 1].set_title('Histogram')
    ax[1, 1].set_xlabel('Runs')
    ax[1, 1].set_ylabel('Frequency')
    ax[1, 1].legend()

    # Show the plot
    

create_subplots()
```

## adding suplots manually

```python
fig = plt.figure(figsize=(10,8))

axs1 = fig.add_subplot(2,2,1)
axs1.plot([2013, 2014, 2015, 2016, 2017],[1000,2000,3000,4000,5000],color='green',marker='o',linestyle='dashdot')

axs2  = fig.add_subplot(2,2,2)
axs2.bar([2013, 2014, 2015, 2016, 2017],[1000,2000,3000,4000,5000],color='violet',edgecolor='black')
axs2.set_xlabel('Years of Experience')
axs2.set_ylabel('Salary')
axs2.legend('Salary Increment')
```
